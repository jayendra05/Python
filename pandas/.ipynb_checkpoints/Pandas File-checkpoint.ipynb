# Merging ()

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'class':[1,2,3,4]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4,5],'class':[5,6,7,8,9]})
df2
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

# Concat()

```python
import pandas as pd
```

```python
sr1=pd.Series([1,2,3,4])
sr1
```

```python
sr2=pd.Series([5,6,7,8])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                  'Name':['A','B','C','D'],
                  'Class':[12,11,12,10]
                 })
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                  'Name':['E','F','G','H'],
                  'Class':[11,12,10,11]
                 
                 })
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],keys=['First df','Second df'])
```

# join

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[10,20,30]})
df1
```

```python
df2=pd.DataFrame({'C':[4,5,6],'D':[40,50,60]})
display(df1,df2)
```

```python
df2.join(df1)
```

```python
df1.join(df2)
```

```python
df1.join(df2,lsuffix='_1')
```
